**Dados**: Dados dos jogadores disponíveis no FIFA 18

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('~/unb/icc/tf/complete.csv')
```

Para analisar a relação entre número de observações e overall (nível de habilidade geral dentro do jogo) dos jogadores, é necessário escolhermos um tipo de gráfico adequado. Como ambas as variáveis são classificadas como quantitativas, podemos utilizar um histograma para estudar a distribuição dos dados. Observando o gráfico abaixo, percebemos uma clara concentração dos dados no intervalo [60,70], o que é esperado. Afinal, poucas pessoas que não são muito boas em futebol tornam-se profissionais, reduzindo o número de jogadores com baixo overall. Além disso, existem poucos jogadores com overall acima de 80, pois poucas pessoas possuem talento o suficiente para alcançar um alto nível no futebol.

```python
plt.hist(df.overall, bins = 16)
plt.xlabel('Nível \"overall"')
plt.ylabel('Frequência')
```

Aqui estamos interessados em estudar a relação entre o valor (Em 100 milhões de Euros) do jogador dentro do FIFA 18 e o seu overall. Nesse caso, usaremos um gráfico de dispersão, pois ambas as variáveis são do tipo quantitativa. Interpretando o gráfico abaixo, percebemos que existe um elevado grau de associação entre as variáveis, pois o gráfico nos demonstra uma curva que nos remete a uma função exponencial.

```python
plt.scatter(df.overall, df.eur_value, c = df.eur_value)
plt.xlabel('Nível \"overall"')
plt.ylabel('Preço (10e8 €)')
```

Sabendo que as ligas nacionais do FIFA 18 é uma variável qualitativa, e o overall médio da liga é classificada como quantitativa, podemos utilizar um gráfico de barras com o intuito de descobrir qual liga possui o maior overall médio. Olhando o gráfico abaixo, nós percebemos que as melhores ligas do mundo (Premier League, Serie A, La Liga, Bundesliga e League 1) são justamente as que possuem o maior overall médio do jogo, conforme o esperado. Além disso, nós também descobrimos que as piores ligas são da Suécia, Brasil e Chile. Selecionamos apenas os 1000 melhores jogadores.

```python
df2 = df.head(1000).groupby(['league'])['overall'].mean()
df2.sort_values().plot.barh()
```
